# Import the matplotlib library, specifically the pyplot module

```python
import matplotlib.pyplot as plt
# https://matplotlib.org/stable/tutorials/index.html#tutorials reference material link
```

## Create the first plot
A simple line connecting the points determined by the x and y coordinates specified by the list

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting the data
plt.plot(x, y)

# Displaying the plot
plt.show()
```

## Add a title and label
Adds labels to the x and y axi, and a title

```python
plt.plot(x, y)

# Adding a title
plt.title('Sample Line Plot')

# Adding labels to the axes
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

## Create a Scatter Plot
"Scatter plots are used to observe relationships between variables. Here's how to create a basic scatter plot:"

```python
x = [5, 2, 9, 4, 7]
y = [10, 5, 8, 4, 2]

# Creating the scatter plot
plt.scatter(x, y)

plt.title('Sample Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

## Make a Bar Chart
"Bar charts are useful for comparisons among discrete categories. Heres how to plot a simple bar chart:"

```python
# Categories and values
categories = ['Red', 'Blue', 'Green', 'Purple']
values = [10, 15, 7, 5]

# Creating the bar chart
plt.bar(categories, values)

plt.title('Simple Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```

## Customizing Colors and Markers
"Changing the color and shape of the markers on a plot can make your data more visually distinct and easier to interpret."

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting with a specific color and marker
plt.plot(x, y, marker='s', color='green')  # 's' stands for square marker

plt.title('Colorful Plot with Square Markers')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

## Line Styles
"You can make a line dashed, dotted, or any other style to differentiate it from other lines or to emphasize certain data."

```python
# Plotting with a dashed line
plt.plot(x, y, linestyle='--', color='red')  # '--' for dashed line

plt.title('Dashed Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

## Adding Text Annotations
Sometimes you might want to draw attention to a specific point or add a text explanation directly on the plot:

```python
# Plotting with a blue line with circle points
plt.plot(x, y, marker='o', color='blue')

# xy = is which data point the arrow points to
# xytext = is the location of the text and starting point of the arrow
# arrowprops = is a dictionary that controls how the arrow looks
plt.annotate('Key Point', xy=(x[2], y[2]), xytext=(x[2]-1, y[2]+3),
             arrowprops=dict(facecolor='black', shrink=0.1),
             fontsize=16, color='red')

plt.title('Plot with Text Annotation')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

## Using Grid Lines
Grid lines can be helpful for better visual measurement and enhancing readability:

```python
# Enabling grid lines
plt.plot(x, y, color='purple')
plt.grid(True)  # Turns on the grid

plt.title('Plot with Grid Lines')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```
